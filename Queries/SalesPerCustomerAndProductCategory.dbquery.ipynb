```sql
USE `databricks-catalog`.gold;

WITH sales_analysis AS (
  SELECT 
  dd.CalendarDate,
  dc.Gender,
  CASE
    WHEN dc.Age >= 18 AND dc.Age <= 24 THEN '18-24'
    WHEN dc.Age >= 25 AND dc.Age <= 34 THEN '25-34'
    WHEN dc.Age >= 35 AND dc.Age <= 44 THEN '35-44'
    WHEN dc.Age >= 45 AND dc.Age <= 54 THEN '45-54'
    WHEN dc.Age >= 55 AND dc.Age <= 64 THEN '55-64'
    WHEN dc.Age >= 65 THEN '65+'
    ELSE '0'
  END AS AgeRange,
  dpc.ProductCategory,
  fs.Quantity,
  fs.UnitPrice,
  fs.TotalAmount
FROM 
  FactSales AS fs
  INNER JOIN DimDate AS dd
    ON fs.CalendarDateId = dd.CalendarDateId
  INNER JOIN DimCustomer AS dc
    ON fs.CustomerId = dc.CustomerId
  INNER JOIN DimProductCategory AS dpc
    ON fs.ProductCategoryId = dpc.ProductCategoryId
)

SELECT 
  CalendarDate,
  Gender,
  AgeRange,
  ProductCategory,
  SUM(Quantity) AS TotalQuantity,
  SUM(UnitPrice) AS TotalUnitPrice,
  SUM(TotalAmount) AS TotalAmount
FROM 
  sales_analysis
GROUP BY 
  CalendarDate,
  Gender,
  AgeRange,
  ProductCategory
ORDER BY 
  CalendarDate DESC,
  ProductCategory

```
